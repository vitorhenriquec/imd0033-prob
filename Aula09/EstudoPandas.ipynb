```python
def which_class(row):
    pclass = row['pclass']
    if pd.isnull(pclass):
        return "Unknown"
    elif pclass == 1:
        return "First Class"
    elif pclass == 2:
        return "Second Class"
    else:
        return "Third Class"
classes = data.apply(which_class, axis=1)
```

```python
classes
```

```python
data["age_cat"] = pd.cut(data.age,
       bins=[0,5,10,18,30,50,65,100],labels=["Infantil","Crianca","Adolescente","Adulto Jovem","Adulto Pleno","Adulto Senior","Idoso"])
#O programa acima faz a mesma coisa do de baixo só que em um linha
```

```python
aux = data.pivot_table(index=["age_cat","survived"],values=["age"], aggfunc="count")
```

```python
aux["age_prob"] = 100*(aux.age/aux.sum().age)
```
